```python
# Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Make plots look cleaner
sns.set()
```

```python
df = pd.read_csv("global_ai_jobs.csv")

# Display first 5 rows
df.head(7)
```

```python
df.dtypes
```

```python
print(df.isna().sum())
```

```python
# checking duplicate rows
df.duplicated()[df.duplicated()==True]
```

```python
print("Shape (rows, columns): ", df.shape,"\n")

print("number of rows: ", df.shape[0])
print("number of columns: ", df.shape[1])
```

```python
# Statistical summary
df.describe(include='all')
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['promotion_speed'], bins=20)
plt.title("Distribution of promotion speed")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['salary_usd'], bins=20)
plt.title("Distribution of Salary")
plt.show()
```

```python
country_salary = df.groupby('country')['salary_usd'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,5))
country_salary.plot(kind='bar')
plt.title("Total salary by Country")
plt.ylabel("Salary")
plt.show()

country_salary
```

```python
job_salary = df.groupby('job_role')['salary_usd'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,5))
job_salary.plot(kind='bar')
plt.title("Total salary by Job role")
plt.ylabel("Salary")
plt.show()

job_salary
```

```python
exper_salary = df.groupby('experience_level')['salary_usd'].sum().sort_values(ascending=False)

exper_salary.head(10)
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='bonus_usd', y='salary_usd', data=df)
plt.title("bonus_usd vs salary_usd")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df[['bonus_usd', 'salary_usd']].corr(), annot=True)
plt.title("Correlation Matrix")
plt.show()
```

```python
# Grouping by Year and calculating the average salary
yearly_salary = df.groupby('year')['salary_usd'].mean()
plt.figure(figsize=(10, 5))
yearly_salary.plot(marker='o', linestyle='-', color='b') # Added markers for clarity
plt.title("Yearly Average Salary Trend in AI")
plt.xlabel("Year")
plt.ylabel("Average Salary (USD)")
plt.grid(True)
plt.show()
```
